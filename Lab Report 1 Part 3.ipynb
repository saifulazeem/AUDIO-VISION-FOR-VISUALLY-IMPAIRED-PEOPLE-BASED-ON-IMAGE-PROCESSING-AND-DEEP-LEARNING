```python
def menu():
    print("Press 1 for Addition")
    print("Press 2 for Subtraction")
    print("Press 3 for Multiplication")
    print("Press 4 for Division")
    print("Press 5 for Factorial")
    print("Press 6 for x power y")
    print("Press 7 for Log")
    print("Press 8 for ln")
    user_choice=int(input("Please Enter Your choice: "))
    if user_choice==1:
        rlst=add()
        print("Resluts: ",rlst)
        continue_program()
    elif user_choice==2:
        rlst=sub()
        print("Resluts: ",rlst)
        continue_program()
    elif user_choice==3:
        rlst=mul()
        print("Resluts: ",rlst)
        continue_program()
    elif user_choice==4:
        rlst=div()
        print("Resluts: ",rlst)
        continue_program()
    elif user_choice==5:
        rlst=fac()
        print("Resluts of Factorial is : ",rlst)
        continue_program()
    elif user_choice==6:
        rlst=power()
        print("Resluts of X power Y is : ",rlst)
        continue_program()
    elif user_choice==7:
        rlst=log()
        print("Resluts of Log is : ",rlst)
        continue_program()
    elif user_choice==8:
        rlst=Ln()
        print("Resluts of Log is : ",rlst)
        continue_program()
    elif user_choice>4 or user_choice!=int:
        print("Please Enter Valid Choice")
        continue_program()
def add():
    x=int(input("Enter Number: "))
    y=int(input("Enter Number: "))
    
    return x+y
def sub():
    x=int(input("Enter Number: "))
    y=int(input("Enter Number: "))
    
    return x-y
def mul():
    x=int(input("Enter Number: "))
    y=int(input("Enter Number: "))
    
    return x*y
def div():
    x=int(input("Enter Number: "))
    y=int(input("Enter Number: "))
    
    return x/y

def fac():
    x=int(input("Enter Number: "))
    i=x-1
    while i>=1:
        rslt=x*i
        x=rslt
        i -=1
    return x

def power():
    x=int(input("Enter Number: "))
    y=int(input("Enter Number: "))
    
    exp=y**x
    return exp

def log():
    x=int(input("Enter Number: "))
    import math
    res=math.log(x)
    return res
def Ln():
    x=int(input("Enter Number: "))
    import numpy as np
    res=np.log(x)
    return res
    
    
    

def continue_program():
    
    print("Press 'c' for Main menu or 'q or any key'  to quit program ")
    req=input("Enter you Choice: ")
    if req=="c":
        menu()
    elif req!="c":
        print("Program Ends")

menu()
```
