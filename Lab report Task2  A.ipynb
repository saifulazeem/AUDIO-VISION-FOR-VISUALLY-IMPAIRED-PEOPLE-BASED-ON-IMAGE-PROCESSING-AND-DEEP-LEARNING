```python
class basic_calc:
    def __init__(self,x,y):
        self.x=x
        self.y=y

    
    def add(self):
        sums=self.x  +self.y
        
    
        return sums
    
    def sub(self):
        subs=self.x-self.y
        #rslt="Subtraction is : "+subs
    
        return subs
    
    def mul(self):
        mul=self.x*self.y
        #rslt="Multiplication is : "+mul
    
        return mul
    
    def div(self):
        divs=self.x/self.y
        #rslt="Division is : "+divs
    
        return divs
        
my_calc=basic_calc(3,3)

print("SUM is : ",my_calc.add())
print("Subtraction is : ",my_calc.sub())
print("Multiplication is : ",my_calc.mul())
print("Division is : ",my_calc.div())

class s_calc(basic_calc): 
    def __init__(self,x): 
        self.x = x 
        self.y = x
    def fac(self):
        i=self.x-1
        while i>=1:
            rslt=self.x*i
            self.x=rslt
            i -=1
        return self.x

    def power(self):
    
    
        exp=self.y**self.x
        return exp

    def log(self):
    
        import math
        res=math.log(self.x)
        return res
    def Ln(self):
    
        import numpy as np
        res=np.log(self.x)
        return res
    def sin_fun(self):
        
        import math
        res=math.sin(self.x)
        return res
    
    def cos_fun(self):
        
        import math
        res=math.cos(self.x)
        return res
    def tan_fun(self):
        
        import math
        res=math.tan(self.x)
        return res
        

my_s_calc=s_calc(3)

print("Factorial is : ",my_s_calc.fac())
print("Power is : ",my_s_calc.power())
print("Log is : ",my_s_calc.log())
print("Natural Log is : ",my_s_calc.Ln())

print("Sin of X is : ",my_s_calc.sin_fun())
print("Cos of X is : ",my_s_calc.cos_fun())
print("Tan of X is : ",my_s_calc.tan_fun())
```
